```python

# This notebook was created to split the AML dataset in 5 Folds for train, test and val 
# Environment: Aixa_ARMED_2
import anndata
from scMEDAL.utils.splitter import DataSplitter
from scMEDAL.utils.utils import read_adata
from config_split_paths import data_path,data2split_foldername,folder_splits
import numpy as np


```

```python
adata2split_path = data_path + "/" + data2split_foldername
print(adata2split_path)
X,var,obs = read_adata(adata2split_path, issparse=False)
adata = anndata.AnnData(X,obs=obs,var=var)
```

```python
adata.obs['celltype'] = adata.obs['celltype'].astype('category')
adata.obs['batch'] = adata.obs["batch"].astype('category')
```

```python
#Check data to split
print("count matrix shape:",adata.shape)
print("n cell types:",len(np.unique(adata.obs["celltype"])))
print("cell types:",np.unique(adata.obs["celltype"]))
print("\nn batches:",len(np.unique(adata.obs["batch"])))
print("batches:",np.unique(adata.obs["batch"]))
```

```python
# Call DataSplitter (5 folds)
splitter = DataSplitter()
# For the Leukemia datase, we will take all donors as seen
DataSplitter.Get_Kfolds_likeAlbert(adata = adata, out_folder=data_path+"/"+folder_splits, n_splits=5, stratified=True, stratify_cols=['batch','celltype'])
```
