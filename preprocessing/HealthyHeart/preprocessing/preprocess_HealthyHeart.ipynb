# Healthy Heart Dataset Processing Script

This script processes the **Healthy Heart** dataset obtained from [Figshare](https://figshare.com/articles/dataset/Batch_Alignment_of_single-cell_transcriptomics_data_using_Deep_Metric_Learning/20499630/2)  
(Yu et al., 2023).

**Dataset Location**  
The dataset should be saved in `scMEDAL_for_scRNAseq/Experiments/data/HealthyHeart_data/raw`.

**Preprocessing and Output**  
The script preprocesses the data and stores the processed output in  
`scMEDAL_for_scRNAseq/Experiments/data/HealthyHeart_data/scenario_id`.  
By default, `scenario_id` is set to `log_transformed_3000hvggenes`.

**Environment**  
This script should be run with the scMEDAL environment.


```python
import os
import sys
import anndata
from pathlib import Path

ROOT_PATH  = Path.cwd().resolve().parents[2]  # two levels up from AML dir (scMEDAL_for_scRNAseq)
sys.path.insert(0, str(ROOT_PATH ))
print(ROOT_PATH )

# Now you can import from the parent directory
from utils.defaults import HH_DATA_DIR,HH_EXPERIMENT_NAME 
from utils.preprocessing import scRNAseq_pipeline_loghvg,H5ADLoader
from utils.utils import save_adata, read_adata
```

```python
HH_DATA_DIR
```

```python
# Path to the directory containing raw data
# Example: /MyscMEDALExpt/Experiments/data/HealthyHeart_data/raw
parent_path = os.path.join(HH_DATA_DIR, "raw")
# file name
healthy_heart_file = "Healthy_human_heart_adata.h5ad"
file_path = os.path.join(parent_path,healthy_heart_file )


# Load adata with H5ADLoader
loader = H5ADLoader(file_path)
loader.load_h5ad()
adata = loader.create_anndata()
# save adata raw
# save_adata(adata,parent_path)
```

```python
adata
```

```python
save_data = False
# expt = "expt_name"
expt = HH_EXPERIMENT_NAME # On this case the HH_EXPERIMENT_NAME has predefined preprocessing 
print("expt:",HH_EXPERIMENT_NAME)
out_path=""
if save_data:
    out_path = os.path.join(HH_DATA_DIR,expt)
    print(out_path)
    # Check if the directory exists, if not, create it
    if not os.path.exists(out_path):
        os.makedirs(out_path)
# Adapt to the type of preprocessing you need
# Log transform log(x+1) +HVG genes
elif expt == "log_transformed_3000hvggenes":
    adata_log_hvg = scRNAseq_pipeline_loghvg(adata, min_genes_per_cell=10, min_cells_per_gene=3, total_counts_per_cell=10000,n_top_genes=3000)
    if save_data:
        save_adata(adata_log_hvg,out_path)
```
